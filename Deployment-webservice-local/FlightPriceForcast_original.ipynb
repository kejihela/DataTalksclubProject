```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
import os
import pickle
```

```python
df= pd.read_csv("dataset/flight_dataset.csv")
df
```

```python
df.info()
```

```python
categorical = ["Airline", "Source", "Destination"]
numerical = ["Total_Stops","Duration_hours","Duration_min"]
```

```python
df = df[categorical + numerical]
df
```

```python
df.Duration_hours = df.Duration_hours *60
df
```

```python
df["duration"] = df["Duration_hours"] + df["Duration_min"]
df
```

```python
target = df["duration"].values
```

```python
df = df.drop(["Duration_hours", "Duration_min", "duration"], axis = 1)
df
```

```python
df = df.to_dict(orient = "records")
```

```python
dv = DictVectorizer()
data_df = dv.fit_transform(df)
```

```python
dv.feature_names_
```

```python
train_df = data_df[:8000]
test_df = data_df[8000:]
y_train =  target[:8000]
y_test = target[8000:]
```

```python
#X_train, y_train, X_test, y_test = train_test_split(data_df,target, test_size=0.25, random_state = 32, shuffle = True)
```

```python
lr = LogisticRegression(solver='liblinear')
lr.fit(train_df, y_train )
```

```python
y_pred = lr.predict(test_df)
```

```python
rmse = mean_squared_error(y_pred, y_test, squared = False)
rmse
```

```python
las = Lasso(0.001, max_iter=10000)
las.fit(train_df, y_train )
```

```python
y_pred = las.predict(test_df)
```

```python
rmse = mean_squared_error(y_pred, y_test, squared = False)
rmse
```

```python
import mlflow

import xgboost as xgb
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from hyperopt.pyll import scope
```

```python

mlflow.set_experiment("Flight prediction")
```

```python
mlflow.set_tracking_uri("sqlite:///mlflow.db")
mlflow.set_tag("Ml engineer", "Adebayo Abdulganiyu Keji")
```

```python
train = xgb.DMatrix(train_df, label=y_train)
valid = xgb.DMatrix(test_df, label=y_test)
```

```python
def objective(params):
    with mlflow.start_run():
        mlflow.set_tag("model", "xgboost")
        mlflow.log_params(params)
        booster = xgb.train(
            params=params,
            dtrain=train,
            num_boost_round=100,
            evals=[(valid, 'validation')],
            early_stopping_rounds=50
        )
        y_pred = booster.predict(valid)
        rmse = mean_squared_error(y_test, y_pred, squared=False)
        mlflow.log_metric("rmse", rmse)

    return {'loss': rmse, 'status': STATUS_OK}
```

```python
search_space = {
    'max_depth': scope.int(hp.quniform('max_depth', 4, 100, 1)),
    'learning_rate': hp.loguniform('learning_rate', -3, 0),
    'reg_alpha': hp.loguniform('reg_alpha', -5, -1),
    'reg_lambda': hp.loguniform('reg_lambda', -6, -1),
    'min_child_weight': hp.loguniform('min_child_weight', -1, 3),
    'objective': 'reg:linear',
    'seed': 42
}

best_result = fmin(
    fn=objective,
    space=search_space,
    algo=tpe.suggest,
    max_evals=6,
    trials=Trials()
)
```

```python
mlflow.end_run()
```
