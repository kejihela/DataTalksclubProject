```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import pandas as pd
import pickle
from sklearn.feature_extraction import DictVectorizer
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/kejihela/DataTalksclubProject/master/dataset/flight_dataset.csv")
df
```

```python
df.info()
```

```python
categorical = ["Airline", "Source", "Destination"]
numerical = ["Total_Stops","Duration_hours","Duration_min"]
```

```python
df = df[categorical + numerical]
df
```

```python
df.Duration_hours = df.Duration_hours *60
df
```

```python
df["duration"] = df["Duration_hours"] + df["Duration_min"]
df
```

```python
df = df.drop(["Duration_hours", "Duration_min", "duration"], axis = 1)
df
```

```python
target = df["duration"].values
```

```python
df = df.to_dict(orient = "records")
```

```python
dv = DictVectorizer()
df = dv.fit_transform(df)
```

```python
train_df = df[:8000]
test_df = df[8000:]
y_train =  target[:8000]
y_test = target[8000:]
```

```python
model = LinearRegression()
model.fit(train_df, y_train)
```

```python
train_pred = model.predict(train_df)

```

```python
mean_absolute_error(train_pred, y_train)
```

```python
val_df = pd.DataFrame()
```

```python
val_pred = model.predict(test_df)
val_df["prediction"] = val_pred
```

```python
mean_absolute_error(val_pred, y_test)
```

```python
with open('models/linear_reg.bin','wb') as f_out:
    pickle.dump(model, f_out)
```

```python
val_df.to_csv("data/reference.csv")
```
